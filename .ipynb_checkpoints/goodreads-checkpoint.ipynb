#### The dataset includes not only books that I've read, but also books that I was reading at that time and books on my reading list. For this analysis, let's look only at the books that I've read.

```python
gr_read = gr[['Title', 'Author', 'My Rating','Average Rating']]
```

#### Let's compare my average rating with the goodreads average rating

```python
my_avg = gr['My Rating'].mean()
my_avg
```

```python
gr['My Rating'].plot.hist()
```

```python
gr_avg = gr['Average Rating'].mean()
gr_avg
```

```python
gr['Average Rating'].plot.hist()
```

##### Clearly, the average goodreads rating is higher than my rating, but it's a difference of only about 0.6 points, goodreads tending towards 4, and me towards 3.

#### Let's see the top-rated books by me and goodreads

```python
top_books = gr[(gr['My Rating']==5) & (gr['Average Rating']>=4.5)]
top_books[['Title', 'Author', 'Author Country', 'My Rating','Average Rating']]
```

##### It's a bit surprising to see only these three books are considered top by me and the goodreads users. It's also worth noting that these books are non-fiction, more specifically about lifestyle: beauty/ portrait photograhy, meditation/ Indian philosophy, and personal health with ayurvedic remedies.
